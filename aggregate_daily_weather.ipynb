```python
df = df1.merge(df2, on = ['POSTE', 'DATE']).drop(['POSTE','FF2M', 'GLOT', 'DIRT'], axis=1)

df['Date'] = pd.to_datetime(df['DATE'].astype(str))
df = df.set_index('Date').drop('DATE', axis=1).rename(
    columns={
        'RR':'P',
        'TN':'Tn',
        'TX':'Tx',
        'TM':'Tm',
        'PMERM':'Pr',
        'INST':'Sun',
        'GRELE':'Hail',
        'ETPMON':'ETPmon',
        'FFM':'Wind'
    })
```
